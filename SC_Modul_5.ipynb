```python
import numpy as np
import matplotlib.pyplot as plt

# Inisialisasi data
data_x = np.array([0, 0.1, 0.2, 0.3, 0.4])
data_y = np.array([0, 0.0819, 0.01341, 0.1646, 0.1797])
delta_x = 0.1

# Menghitung turunan pertama menggunakan metode selisih maju di titik x = 0
turunan_maju = (data_y[1] - data_y[0]) / delta_x

# Menghitung turunan pertama menggunakan metode selisih tengah di titik x = 0.2
turunan_tengah = (data_y[3] - data_y[1]) / (2 * delta_x)

# Menampilkan hasil
print(f"Hasil turunan di x=0 (metode maju): {turunan_maju}")
print(f"Hasil turunan di x=0.2 (metode tengah): {turunan_tengah}")
```

```python
import numpy as np

# Definisi fungsi yang akan diintegrasikan
def fungsi(x):
    return np.sin(x)

batas_awal = 0
batas_akhir = np.pi

# Metode Trapesium dengan 8 panel
jumlah_panel_8 = 8
lebar_8 = (batas_akhir - batas_awal) / jumlah_panel_8
titik_x_8 = np.linspace(batas_awal, batas_akhir, jumlah_panel_8 + 1)
nilai_y_8 = fungsi(titik_x_8)
hasil_trap_8 = lebar_8 * (nilai_y_8[0] + 2 * np.sum(nilai_y_8[1:-1]) + nilai_y_8[-1]) / 2
print(f"Hasil integral trapesium (8 panel): {hasil_trap_8}")

# Metode Trapesium dengan 16 panel
jumlah_panel_16 = 16
lebar_16 = (batas_akhir - batas_awal) / jumlah_panel_16
titik_x_16 = np.linspace(batas_awal, batas_akhir, jumlah_panel_16 + 1)
nilai_y_16 = fungsi(titik_x_16)
hasil_trap_16 = lebar_16 * (nilai_y_16[0] + 2 * np.sum(nilai_y_16[1:-1]) + nilai_y_16[-1]) / 2
print(f"Hasil integral trapesium (16 panel): {hasil_trap_16}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi yang akan diintegrasikan
def fungsi_sin(x):
    return np.sin(x)

awal = 0
akhir = np.pi
nilai_tepat = 2
daftar_error = []
jumlah_panel = [2, 4, 8, 16, 32, 64, 128]

for panel in jumlah_panel:
    langkah = (akhir - awal) / panel
    titik_x = np.linspace(awal, akhir, panel + 1)
    titik_y = fungsi_sin(titik_x)
    hasil_integral = langkah * (titik_y[0] + 2 * np.sum(titik_y[1:-1]) + titik_y[-1]) / 2
    galat = abs(nilai_tepat - hasil_integral)
    daftar_error.append(galat)
    print(f"Panel = {panel}, Nilai Integral = {hasil_integral}, Galat = {galat}")

# Visualisasi error terhadap jumlah panel (log-log plot)
plt.figure(figsize=(10, 6))
plt.plot(jumlah_panel, daftar_error, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Jumlah Panel (log skala)')
plt.ylabel('Galat (log skala)')
plt.title('Galat pada Metode Trapesium terhadap Jumlah Panel')
plt.grid(True, which='both')
plt.show()
```
