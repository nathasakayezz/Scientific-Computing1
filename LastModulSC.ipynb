```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

# Definisi fungsi turunan
def fungsi_turunan(waktu, nilai):
    return np.exp(-waktu)

# Parameter metode Euler
delta_t = 0.1
waktu_awal = 0
waktu_akhir = 1

# Vektor waktu
waktu = np.arange(waktu_awal, waktu_akhir + delta_t, delta_t)

# Nilai awal
nilai_awal = -1
solusi_numerik = np.zeros(len(waktu))
solusi_numerik[0] = nilai_awal

# Metode Euler
for indeks in range(len(waktu) - 1):
    solusi_numerik[indeks + 1] = solusi_numerik[indeks] + delta_t * fungsi_turunan(waktu[indeks], solusi_numerik[indeks])

# Plot hasil
plt.figure(figsize=(12, 8))
plt.plot(waktu, solusi_numerik, 'ro--', label='Euler (t=0.1)')  # Merah
plt.plot(waktu, -np.exp(-waktu), color='gold', linewidth=2, label='Solusi Eksak')  # Kuning
plt.xlabel('Waktu')
plt.ylabel('Nilai')
plt.title('Pendekatan Euler vs Solusi Eksak')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi turunan dari persamaan diferensial
def turunan(x, y):
    return x**2 - 4*y

# Parameter untuk metode numerik
delta_x = 0.1
batas_awal = 0
batas_akhir = 1

# Vektor nilai x dan inisialisasi y
x_vals = np.arange(batas_awal, batas_akhir + delta_x, delta_x)
y_awal = 1
estimasi_y = np.zeros(len(x_vals))
estimasi_y[0] = y_awal

# Metode Euler
for j in range(len(x_vals) - 1):
    estimasi_y[j + 1] = estimasi_y[j] + delta_x * turunan(x_vals[j], estimasi_y[j])

# Solusi eksak
def solusi_analitik(x):
    return (x**2 / 4 - x / 8 + 1 / 32) + (31 / 32) * np.exp(-4 * x)

# Visualisasi hasil
plt.figure(figsize=(12, 8))
plt.plot(x_vals, estimasi_y, 'ro--', label='Euler (x=0.1)')
plt.plot(x_vals, solusi_analitik(x_vals), color='gold', linewidth=2, label='Solusi Eksak')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Penyelesaian $y\' + 4y = x^2$ dengan Metode Euler')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definisi fungsi turunan
def fungsi_ode(t, y):
    return np.sin(y)

# Parameter simulasi
step = 0.1
awal = 0
akhir = 5
nilai_awal = 1

# Vektor waktu dan inisialisasi solusi
titik_waktu = np.arange(awal, akhir + step, step)
aproksimasi = np.zeros(len(titik_waktu))
aproksimasi[0] = nilai_awal

# Metode Euler eksplisit
for idx in range(len(titik_waktu) - 1):
    aproksimasi[idx + 1] = aproksimasi[idx] + step * fungsi_ode(titik_waktu[idx], aproksimasi[idx])

# Solusi numerik akurat menggunakan solve_ivp
solusi_ivp = solve_ivp(fungsi_ode, [awal, akhir], [nilai_awal], t_eval=titik_waktu)

# Visualisasi hasil
plt.figure(figsize=(12, 8))
plt.plot(titik_waktu, aproksimasi, 'ro--', label='Euler (t=0.1)')
plt.plot(titik_waktu, solusi_ivp.y[0], color='orange', label='solve_ivp')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Perbandingan Euler dan solve_ivp pada $y\' = \sin y$')
plt.legend()
plt.grid(True)
plt.show()
```
