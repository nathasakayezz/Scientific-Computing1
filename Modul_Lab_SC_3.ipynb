```python
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot') #style gambar grafik
```

MODUL 3

Olympiade tahun 1990, juara lomba sprint 100 m adalah Ben Johnson. Catatan larinya ketika 0 detik kecepatannya adalah 10m/detik. Pada saat 5 meter kecepatannya adalah 5m/detik. Pada saat 10 meter kecepatannya adalah 20m/detik. Berapa kecepatan larinya saat pada jarak 7.2m? Jawablah menggunakan coding python linier interpolation di bawah.

```python
# Data jarak dan kecepatan
nilai_jarak = np.array([0, 5, 10])
nilai_kecepatan = np.array([10, 15, 20])

# Membuat fungsi interpolasi linear
fungsi_interpolasi = interp1d(nilai_jarak, nilai_kecepatan)

# Titik target untuk interpolasi
target_jarak = 7.2

# Membuat nilai interpolasi untuk visualisasi
jarak_interp = np.linspace(min(nilai_jarak), max(nilai_jarak), 100)
kecepatan_interp = fungsi_interpolasi(jarak_interp)

# Melakukan prediksi
hasil_prediksi = fungsi_interpolasi(target_jarak)
print(hasil_prediksi)

# Membuat plot
plt.figure(figsize=(10, 8))
plt.plot(nilai_jarak, nilai_kecepatan, '-ob', label="Data Awal")
plt.plot(jarak_interp, kecepatan_interp, 'b', label="Interpolasi Linear")
plt.plot(target_jarak, hasil_prediksi, 'ro', label=f"Prediksi di {target_jarak}")
plt.title('Interpolasi Linear pada x = 7.2')
plt.xlabel('Jarak')
plt.ylabel('Kecepatan')
plt.legend()
plt.show()
```
